```python
# Initialize lists to store data
drivers = []
year = []
australia = []
styria=[]
malaysia = []
china = []
turkey = []
bahrain=[]
spain = []
monaco = []
canada = []
france = []
azerbaijan=[]
austria=[]
europe = []
britain = []
silverstone = []
germany = []
hungary = []
belgium = []
netherlands = []
italy = []
tuscany = []
singapore = []
japan = []
russia=[]
eifel=[]
portugal=[]
emilia_romagna=[]
sakhir=[]
south_korea = []
india=[]
abu_dhabi=[]
qatar=[]
saudi_arabia=[]
usa=[]
las_vegas=[]
miami=[]
mexico=[]  
brazil=[]
total_points = []


# Example header for requests
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Iterate through all URLs in the list
for url in url2000:  # Replace `url2011` with your list of URLs
    # Extract the year from the URL (assuming the format is like https://www.statsf1.com/en/1950.aspx)
    year_from_url = url.split('/')[-1].split('.')[0]

    # Fetch and parse the page content
    response = requests.get(url, headers=header)
    sel = Selector(response)
    container = sel.css("table#ctl00_CPH_Main_TBL_CHP_Drv tr")

    # Extract race data for each race column
    all_data = sel.css("table#ctl00_CPH_Main_TBL_CHP_Drv td::text").getall()
    Australia = all_data[3::21]
    Malaysia = all_data[4::21]
    China = all_data[5::21]
    Turkey = all_data[6::21]
    Spain = all_data[7::21]
    Monaco = all_data[8::21]
    Canada = all_data[9::21]
    Europe = all_data[10::21]
    Britain = all_data[11::21]
    Germany = all_data[12::21]
    Hungary = all_data[13::21]
    Belgium = all_data[14::21]
    Italy = all_data[15::21]
    Singapore = all_data[16::21]
    Japan = all_data[17::21]
    South_Korea = all_data[18::21]
    India = all_data[19::21]
    Abu_Dhabi = all_data[20::21]
    Brazil = all_data[21::21]
    total_points_data = all_data[22::21]
    Styria=[]
    Turkey=[]
    Bahrain=[]  
    France=[]
    Azerbaijan=[]
    Austria=[]
    Silverstone=[]
    Netherlands=[]
    Tuscany=[]
    Russia=[]
    Eifel=[]
    Portugal=[]
    Emilia_romagna=[]
    Sakhir=[]
    Qatar=[]
    Saudi_arabia=[]
    USA=[]
    Las_Vegas=[]
    Miami=[]
    Mexico=[]

    # Generate a list of years using the extracted year
    Year = [year_from_url] * (len(container) - 1)  # Exclude header row

    # Updated Code: Handle missing data and append zeros if necessary

# Ensure the minimum length is calculated for all lists involved
max_length = min(
    len(container) - 1, len(Australia), len(total_points_data)
)

# Loop through each container to extract data (skip the first row, index 0)
for i in range(1, max_length + 1):
    # Extract driver name and fallback options
    driver_name = container[i].css("td a::text").get()
    world_champion = container[i].css("td span.WorldChp::text").get()
    champ = container[i].css("td span.CurChpDriver::text").get()
    driver1 = container[i].css("td span.CurDriver::text").get()

    # Append the driver name or fallback options
    drivers.append(driver_name if driver_name else (world_champion if world_champion else (champ if champ else driver1)))

    # Handle missing data by appending zero for races with no data
    australia.append(Australia[i - 1] if i - 1 < len(Australia) else "0")
    hungary.append(Hungary[i - 1] if i - 1 < len(Hungary) else "0")
    britain.append(Britain[i - 1] if i - 1 < len(Britain) else "0")
    spain.append(Spain[i - 1] if i - 1 < len(Spain) else "0")
    belgium.append(Belgium[i - 1] if i - 1 < len(Belgium) else "0")
    italy.append(Italy[i - 1] if i - 1 < len(Italy) else "0")
    monaco.append(Monaco[i - 1] if i - 1 < len(Monaco) else "0")
    canada.append(Canada[i - 1] if i - 1 < len(Canada) else "0")
    europe.append(Europe[i - 1] if i - 1 < len(Europe) else "0")
    germany.append(Germany[i - 1] if i - 1 < len(Germany) else "0")
    singapore.append(Singapore[i - 1] if i - 1 < len(Singapore) else "0")
    japan.append(Japan[i - 1] if i - 1 < len(Japan) else "0")
    south_korea.append(South_Korea[i - 1] if i - 1 < len(South_Korea) else "0")
    india.append(India[i - 1] if i - 1 < len(India) else "0")
    abu_dhabi.append(Abu_Dhabi[i - 1] if i - 1 < len(Abu_Dhabi) else "0")
    brazil.append(Brazil[i - 1] if i - 1 < len(Brazil) else "0")
    malaysia.append(Malaysia[i - 1] if i - 1 < len(Malaysia) else "0")
    china.append(China[i - 1] if i - 1 < len(China) else "0")
    turkey.append(Turkey[i - 1] if i - 1 < len(Turkey) else "0")
    styria.append(Styria[i - 1] if i - 1 < len(Styria) else "0")
    bahrain.append(Bahrain[i - 1] if i - 1 < len(Bahrain) else "0")
    france.append(France[i - 1] if i - 1 < len(France) else "0")
    azerbaijan.append(Azerbaijan[i - 1] if i - 1 < len(Azerbaijan) else "0")
    austria.append(Austria[i - 1] if i - 1 < len(Austria) else "0")
    silverstone.append(Silverstone[i - 1] if i - 1 < len(Silverstone) else "0")
    netherlands.append(Netherlands[i - 1] if i - 1 < len(Netherlands) else "0")
    tuscany.append(Tuscany[i - 1] if i - 1 < len(Tuscany) else "0")
    russia.append(Russia[i - 1] if i - 1 < len(Russia) else "0")
    eifel.append(Eifel[i - 1] if i - 1 < len(Eifel) else "0")
    portugal.append(Portugal[i - 1] if i - 1 < len(Portugal) else "0")
    emilia_romagna.append(Emilia_romagna[i - 1] if i - 1 < len(Emilia_romagna) else "0")
    sakhir.append(Sakhir[i - 1] if i - 1 < len(Sakhir) else "0")
    qatar.append(Qatar[i - 1] if i - 1 < len(Qatar) else "0")
    saudi_arabia.append(Saudi_arabia[i - 1] if i - 1 < len(Saudi_arabia) else "0")
    usa.append(USA[i - 1] if i - 1 < len(USA) else "0")
    las_vegas.append(Las_Vegas[i - 1] if i - 1 < len(Las_Vegas) else "0")
    miami.append(Miami[i - 1] if i - 1 < len(Miami) else "0")
    mexico.append(Mexico[i - 1] if i - 1 < len(Mexico) else "0")
    total_points.append(total_points_data[i - 1] if i - 1 < len(total_points_data) else "0")

    # Append the year
    year.append(Year[i - 1])

# Ensure the `year` list matches the length of other data
year = year[:len(drivers)]

# Final data structure is ready for use
print("Data extraction complete!")
print(f"Number of drivers: {len(drivers)}")
print(f"Last driver: {drivers[-1] if drivers else 'No data'}")
```

```python


# Calculate quartiles for each driver
driver_stats = amul.groupby("Drivers")["Total Points"].describe(percentiles=[0.25, 0.5, 0.75])
driver_stats = driver_stats.reset_index()[["Drivers", "25%", "50%", "75%"]]
driver_stats.columns = ["Drivers", "Q1", "Median", "Q3"]

# Get the latest year for each driver
latest_years = amul.groupby("Drivers")["Year"].max().reset_index()

# Merge to get the latest year with driver stats
driver_stats = driver_stats.merge(latest_years, on="Drivers")

# Select top 10 drivers based on median points
top_10_drivers = driver_stats.sort_values(by="Median", ascending=False).head(10)

# Reshape data for Plotly (to show Q1, Median, and Q3 as separate points)
top_10_drivers_melted = top_10_drivers.melt(id_vars=["Drivers", "Year"], value_vars=["Q1", "Median", "Q3"],
                                            var_name="Statistic", value_name="Points")

# Create an interactive scatter plot with tooltips including the year
fig = px.scatter(
    top_10_drivers_melted,
    x="Points",
    y="Drivers",
    color="Statistic",
    symbol="Statistic",
    labels={"Points": "Total Points", "Drivers": "Top 10 Drivers", "Statistic": "Quartile"},
    title="Top 10 Drivers - Total Points Distribution Across Multiple Seasons",
    template="plotly_white",
    hover_data={"Statistic": True, "Points": ":.2f", "Year": True}  # Include year in tooltip
)

# Add a line range to indicate Q1-Q3 range
for driver in top_10_drivers["Drivers"]:
    q1 = top_10_drivers.loc[top_10_drivers["Drivers"] == driver, "Q1"].values[0]
    q3 = top_10_drivers.loc[top_10_drivers["Drivers"] == driver, "Q3"].values[0]
    year = top_10_drivers.loc[top_10_drivers["Drivers"] == driver, "Year"].values[0]

    fig.add_shape(
        type="line",
        x0=q1, x1=q3,
        y0=driver, y1=driver,
        line=dict(color="gray", width=2),
    )

# Customize layout and tooltip settings
fig.update_traces(marker=dict(size=10, line=dict(width=1, color="black")))
fig.update_layout(
    xaxis_title="Total Points",
    yaxis_title="Drivers",
    hovermode="closest",
    showlegend=True,
)

# Show the interactive plot
fig.show()
```
